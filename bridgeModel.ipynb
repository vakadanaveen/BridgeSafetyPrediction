```python
#import io
import pandas as pd
df=pd.read_csv('lasyadata.csv')
```

```python
x=df.iloc[:,:2]
```

```python
yd = df.iloc[:,2]
```

```python
x.shape
```

```python
x_train = x[:170]
yd_train = yd[:170]
```

```python
x_train
```

```python
yd_train
```

```python
reg = LinearRegression().fit(x_train, yd_train)
```

```python
x_test = x[171:]
yd_test = yd[171:]
```

```python
reg.predict(x_test)
```

```python
yd_test
```

```python
from sklearn.externals import joblib 
  
# Save the model as a pickle in a file 
joblib.dump(reg, 'model.pkl') 
```

```python

from sklearn.externals import joblib 
# Load the model from the file 
reg_from_joblib = joblib.load('model.pkl')  
 
LOCATION = "Arizona" #Give your location here
d={
    "Alabama": 84000,
    "Alaska" : 145000,
    "Arizona": 80000
}
# Use the loaded model to make predictions 
reg_from_joblib.predict(x_test) 
test_yourself = np.array([[40,7]]) # Give your value here
pv = reg_from_joblib.predict(test_yourself)
if (pv > d[LOCATION]):
  print("Not safe")
else:
  print("You are safe to drive in "+LOCATION)
```
